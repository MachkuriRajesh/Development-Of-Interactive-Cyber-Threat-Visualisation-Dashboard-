**TITLE**:Stock Market Trend Analysis and Prediction

**Description**: Stock Market Trend Analysis and Prediction is the process of studying historical stock market data—such as prices, volume, and market sentiment—to identify patterns and forecast future price movements. It combines data analysis, statistics, and AI/ML techniques to support informed investment and trading decisions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create dummy stock market dataset
np.random.seed(42)

dates = pd.date_range(start="2015-01-01", end="2025-01-01", freq="B")
n = len(dates)

price = np.cumsum(np.random.normal(0.2, 2, n)) + 100
open_price = price + np.random.normal(0, 1, n)
close_price = price + np.random.normal(0, 1, n)
high_price = np.maximum(open_price, close_price) + np.random.uniform(0, 2, n)
low_price = np.minimum(open_price, close_price) - np.random.uniform(0, 2, n)
volume = np.random.randint(1_000_000, 10_000_000, n)
sentiment = np.random.uniform(-1, 1, n)

df = pd.DataFrame({
    "Date": dates,
    "Open": open_price,
    "High": high_price,
    "Low": low_price,
    "Close": close_price,
    "Volume": volume,
    "Market_Sentiment": sentiment
})

# Save dataset
df.to_csv("stock_market_ai_dummy_dataset.csv", index=False)
print("HELLO WORLD")
print("DataSet Imported Sucessfully")
```

```python
# Closing Price Trend
plt.figure()
plt.plot(df["Date"], df["Close"])
plt.title("Stock Closing Price Trend (2015–2025)")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python

# Trading Volume Trend
plt.figure()
plt.plot(df["Date"], df["Volume"])
plt.title("Trading Volume Trend")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python



# Market Sentiment vs Close Price
plt.figure()
plt.scatter(df["Market_Sentiment"], df["Close"])
plt.title("Market Sentiment vs Close Price")
plt.xlabel("Market Sentiment")
plt.ylabel("Close Price")
plt.show()
```
